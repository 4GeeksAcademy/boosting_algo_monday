```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/decision-tree-project-tutorial/main/diabetes.csv')
```

```python
X = main_df.drop(columns='Outcome')
y = main_df['Outcome']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
grad_boost_model = GradientBoostingClassifier(learning_rate=5, n_estimators=100)
```

```python
grad_boost_model.fit(X_train, y_train)
```

```python
y_preds = grad_boost_model.predict(X_train)
```

```python
y_test_preds = grad_boost_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_true = y_train, y_pred=y_preds)
```

```python
accuracy_score(y_true=y_train, y_pred=y_preds)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyperparameters = {
    "learning_rate" : [0.1, 0.2, 0.3, 0.4], 
    "n_estimators" : [20, 50, 100, 150], 
    "max_depth" : [2, 5, 8, 10],
}
```

```python
grid_model = GridSearchCV(GradientBoostingClassifier(random_state = 25), param_grid = hyperparameters, scoring="accuracy", cv = 5)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
best_grad_model = GradientBoostingClassifier(learning_rate=0.2, max_depth=2, n_estimators=20, random_state=25)
```

```python
best_grad_model.fit(X_train, y_train)
```

```python
y_train_preds = best_grad_model.predict(X_train)
```

```python
y_test_preds = best_grad_model.predict(X_test)
```

```python
accuracy_score(y_true=y_train, y_pred=y_train_preds)
```
